##  example dataframe

```python
import pandas as pd
```

```python
data = {
    "NAME" : ["max","john","alice"],
    "AGE" : [23,22,34],
    "COUNTRY" : ["USA","Canada","UK"]
}
```

```python
df = pd.DataFrame(data)

print(df)
```

```python
new_row = pd.DataFrame([{   
    "NAME" : "bob",
    "AGE" : 23,
    "COUNTRY" : "india"}])

df = pd.concat([df,new_row],ignore_index = True)

print(df)
```

```python
df.loc[df['NAME']=='max','AGE']=24

print(df)
```

## Accessing & Filtering Data

```python
df['Quantity'].head()  # single columns
```

### Conditional Filtering (Boolean masks)

```python
df[df['Quantity'] >= 4]
```

```python
df[df['Category']=='Electronics'].head()
```

```python
df[(df['Quantity'] >= 4) & (df['Country'] == 'Canada')] 
```

### Advanced Filtering

```python
df[df['CustomerName'].str.startswith('A')] 
```

```python
df[df['Country'].isin(['USA', 'UK'])]
```

```python
df[~df['Country'].isin(['Canada'])].head()
```

```python
new_row = pd.DataFrame([{
    "OrderID"      : 1001,
    "CustomerName" : "Arul",
    "Product"      : "Laptop",
    "Category"     : "Electronics",
    "Quantity"     : 2,
    "Price"        : 75000,
    "OrderDate"    : "2024-06-07",
    "Shipped"      : "Yes",
    "Country"      : "India"
}])

df = pd.concat([df,new_row], ignore_index = True)

df.to_csv("orders.csv", index=False)

print(df.tail())
```
