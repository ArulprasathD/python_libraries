#   CRUD Operations in Pandas 

#### create(insert row)

```python
new_row = pd.DataFrame([{
    	'SaleID' : '10',
        'CustomerName' : 'joe',
        'Product' : 'Mouse',
        'Category' : 'Accessories',
        'Quantity' : 3,
        'UnitPrice' : 23000,
        'SaleDate' : '2025-01-22',
        'City': 'Chennai'
}])

df = pd.concat([df,new_row],ignore_index = True)

df.to_csv('retail_sales_dataset.csv',index = False)

df
```

####  Read (Access Data)

```python
df.loc[df['CustomerName'] == 'joe']
```

```python
df.iloc[10]
```

####  Update (Modify Values)

```python
df.loc[df['CustomerName'] == 'joe','SaleID'] = 2011

df.tail()
```

```python
df['City'] = df['City'].str.upper()

df
```

```python
df.duplicated().sum() 
```

```python
df = df.drop_duplicates()
df
```

####  Delete (Drop Data)

```python
df = df.drop(8,axis=0)

df
```

```python
df = df.drop('City', axis=1,errors='ignore')   

df
```

```python
df.isnull().sum()
```
