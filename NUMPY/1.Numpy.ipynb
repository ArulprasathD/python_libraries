```python
#1d

arr = np.array([10,20,30])

print("one dimensional array :",arr)
```

```python
#2d

arr2d = np.array([[10,20,30],[40.0,50,60]])

print("two dimensional array : \n",arr2d)
```

#  type

```python
print("Type : ",type(arr))
```

```python
print("Type : ",type(arr2d))
```

# shape

```python
print(arr.shape)
```

```python
print(arr2d.shape)
```

#   size

```python
print(arr.size)
print(arr2d.size)
```

#  find dimensional

```python
print("dimensional array? : ",arr.ndim)
print("dimensional array? : ",arr2d.ndim)
```

#   Data type


```python
print(arr.dtype)
print(arr2d.dtype)
print(arr2d)
```

```python
arr2d = np.array([[10,20,30],[40.0,50,60]],dtype = int)   #explicit

print(arr2d.dtype)
print(arr2d)
```
