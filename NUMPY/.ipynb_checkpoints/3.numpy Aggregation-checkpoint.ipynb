```python
import numpy as np

arr = np.array([1,2,3,4,5])
```

```python
print("SUM : ",np.sum(arr))
```

```python
print("MEAN :",np.mean(arr))
```

```python
print("MEDIAN : ",np.median(arr))
```

```python
print("MIN : ",np.min(arr))
```

```python
print("MAX : ",np.max(arr))
```

```python
#2D


arr = np.array([[1,2,3],[4,5,6]])

```

```python
print("SUM : ",np.sum(arr))
```

```python
print("sum of coloumn : ",np.sum(arr,axis = 0))   #axis = 0 is coloumn
```

```python
print("sum of coloumn : ",np.sum(arr,axis = 1))   #axis =1 is row
```

```python
print("sum of value greater that 3 : ",np.sum(arr[arr > 3]))   #condition 
```

```python
print("sum of greater that 2 : ",np.sum(arr[arr > 2]))   #condition 
```
